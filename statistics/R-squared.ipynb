# R-squared

R-squared is a metric of correlation (*[-1,1] are good values for correlation, it tells two variables are strongly related*).
It is the proportion of variation explained by the relationship of two variables.
It's range is from 0 to 1 or 0 to 100%.

- a small value, means that the variation in the data isn't very explained by the relationship between the variables.
- a big value means a good relationship.





```python
import numpy as np
import seaborn as sns
from sklearn.linear_model import LinearRegression
```

```python
x = np.arange(1,10,0.3)
y = np.arange(11,20,0.3)
a = [x,y]
sns.scatterplot( x=x, y=y)
sns.regplot(x=x, y=y);
```

## Linear regression

Let's use a linear regression to plot the predicted value

```python
model = LinearRegression()
x.reshape(-1, 1)
#y.reshape(-1, 1)
model.fit(x.reshape(-1, 1), y)
r_squared = model.score(x.reshape(-1, 1), y)
print(f'This is a perfect example, so r squared must be one, and it is = {r_squared}')
```

```python
sns.scatterplot( x=x, y=y)
sns.regplot(x=x, y=model.predict(x.reshape(-1,1)))
```

### Let's add some noise in the data


```python
rng = np.random.RandomState(0) 

x = np.arange(1,10,0.3)
y = np.arange(11,20,0.3)
isPositive = False
print(y)
for i in range(len(y)):
    if isPositive:
        y[i]+= rng.random()*100
    else:
        y[i]-= rng.random()*100
    isPositive = !isPositive
print(y)
a = [x,y]
sns.scatterplot( x=x, y=y)
sns.regplot(x=x, y=y);
```

```python
model = LinearRegression()
x.reshape(-1, 1)
#y.reshape(-1, 1)
model.fit(x.reshape(-1, 1), y)
r_squared = model.score(x.reshape(-1, 1), y)
print(f'This is a extreme case, so our r-squared is low = {r_squared}')
```

```python
sns.scatterplot( x=x, y=y)
sns.regplot(x=x, y=model.predict(x.reshape(-1,1)),scatter_kws={"color": "green"}, line_kws={"color": "red"})
```

In the graph above, we can see the difference between the predicted (green and red), and the actual value.
