# The Central Limit Theorem


Three important things to know before learning about The central limit theorem are:

- The normal distribution (we have a explanation about it)
- Sampling from a distribution


```python
import numpy as np
import seaborn as sns
sns.set(rc={'figure.figsize':(11.7,8.27)})
import matplotlib.pyplot as plt
```

```python
rng = np.random.RandomState(0) 
```

```python
s = rng.uniform(-1,0,100000)
```

```python
ax = sns.histplot(s, kde=True, color="blue")
```

```python
n_samples = 3000
means = []
for i in range(n_samples):
    s = rng.uniform(-1,0,100000)
    means.append(np.mean(s))
```

```python
ax = sns.histplot(means, kde=True, color="blue")
```

Let's understand what is happening. The first plot represents one sampling from a uniform distribution.
When we have a big number of samples, and we get the mean of each sample, whe can plot a new histogram, and tã dã, we have a normal distribution.

Im **most** cases, it doesn't matter which is the distribution that a sample came from, because the mean of several samples will be a normal distribution.
The **most** case, because to use this theorem (Cental Limit Theorem, by the way), we have to be able to calculate a mean from the sample.
Extra study: the Cauchy distribution (it doesn't have a sample mean)
