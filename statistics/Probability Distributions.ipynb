# Probability distributions

It is a statistical function that describes a phenomenon.
It shows how probabilities of measurements are distributed.

Some examples are:

    - Normal Distribution: High/low parts show where measurements are more/less likely

```python
import numpy as np
import seaborn as sns
sns.set(rc={'figure.figsize':(11.7,8.27)})
```

```python
rng = np.random.RandomState(0) 
```

## Normal distribution

The y-axis representes the relative probability of observing the phenomenon.
For example, in the figure below, 0 is more likely  to be observed and -2 or 2 is less likely.

```python
samples_size=[5,50,500,5000]
for i in samples_size:
    normal=rng.normal(
        loc=0.0,        # The mean of the distribution
        scale=1.0,      # The standard deviation 
        size=i       # The size or shape of your array
    )
    sns.histplot(normal, kde=True)
```

The width of a curve is defined by the standard deviation. 

```python
samples_size=[5000]
for i in samples_size:
    step=0.1
    normal=rng.normal(
        loc=0.0,        # The mean of the distribution
        scale=0.4,      # The standard deviation 
        size=i       # The size or shape of your array
    )
    sns.histplot(normal, kde=False, color="lime",bins=np.arange(-4.0,5.0,0.1))
    normal=rng.normal(
        loc=0.0,        # The mean of the distribution
        scale=1.0,      # The standard deviation 
        size=i       # The size or shape of your array
    )
    sns.histplot(normal, kde=False, color="blue",bins=np.arange(-4.0,5.0,0.1))
```
