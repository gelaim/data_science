# Bootstrapping
Replicating a experiment several times is expensive, instead of doing that we can use bootstrapping.

In the bootstrapping we select data from a dataset with replacement. This is called **sampling with replacement**.


Bootstrapping consists of 4 steps:

1. Make a Bootstrapped dataset
2. Calculate something the mean, std, median)
3. Keep track of the calculation of step 2
4. repeat it several times

```python
import numpy as np
import seaborn as sns
sns.set(rc={'figure.figsize':(11.7,8.27)})
import matplotlib.pyplot as plt
rng = np.random.RandomState(0) 
```

```python
population_size = 5000

population_distribution = rng.normal(
    loc=2.0,        # The mean of the distribution
    scale=3,      # The standard deviation 
    size=population_size      # The size or shape of your array
)
```

```python
ax = sns.histplot(population_distribution, kde=True, color="blue",bins=np.arange(-10.0,15.0,0.5))

q3,q1 = np.percentile(population_distribution,[95,5])
plt.axvline(q1, color='red')
plt.axvline(q3, color='red')
```

```python
sample_distribution = np.random.choice(population_distribution, population_size//20, replace=False)
```

```python
ax = sns.histplot(sample_distribution, kde=True, color="blue",bins=np.arange(-10.0,15.0,0.5))

q3,q1 = np.percentile(sample_distribution,[95,5])
plt.axvline(q1, color='red')
plt.axvline(q3, color='red')
```

```python
bootstrapp_samples = 50000
my_samples = []
for _ in range(bootstrapp_samples):
    x = np.random.choice(sample_distribution, size=len(sample_distribution), replace=True)
    my_samples.append(x.mean())
```

```python
ax = sns.histplot(my_samples, kde=True, color="blue",bins=np.arange(-0.0,5.0,0.1))
```

```python
print(f'The mean of the bootstrapping is {np.mean(my_samples)}')
print(f'the standard deviation of the bootstrapping mean is {np.std(my_samples, ddof=1)}. And this is close to the actual mean! We will study this in other moment.')
```

```python
print(f'The mean of the sample distribution is {np.mean(sample_distribution)}')
print(f'the standard deviation of the sample is {np.std(sample_distribution, ddof=1)}')
```

```python
print(f'The mean of the population distribution is {np.mean(population_distribution)}')

print(f'the standard deviation of the population is {np.std(population_distribution, ddof=0)}')
```
