# Histograms

- Is a graphical representation to visualize numerical data

- Is organized in bins: don't rely on the default settings


- We can use a histogram to predict the probability of getting future measurements
    - Normal distribution
    - exponential distribution


### Dataset Example 

```python
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
data = load_iris()
df = pd.DataFrame(data=data.data, columns=data.feature_names)
```

```python
df.hist(column='sepal length (cm)',bins=3)
```

```python
df.hist(column='sepal length (cm)',bins=10)
```

```python
df.hist(column='sepal length (cm)',bins=30)
```

### Numpy Example

```python
rng = np.random.RandomState(0) 
```

```python
a = np.hstack((rng.normal(size=1000),
               rng.normal(loc=5, scale=2, size=1000)))
```

```python
a
```

```python
_ = plt.hist(a, bins=5)
```

```python
_ = plt.hist(a, bins=15)
```

```python
a = np.hstack((rng.exponential(size=1000),
               rng.exponential(loc=5, scale=2, size=1000)))
```
