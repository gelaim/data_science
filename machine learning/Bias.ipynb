# Bias


Bias is the inability for a machine learning method to capture the true relationship of a model.

To understand it, lets create a gaussian-like dataset

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(rc={'figure.figsize':(11.7,8.27)})
rng = np.random.RandomState(0) 
x = np.linspace(0, 0.7 * np.pi, num=50)

n = np.random.normal(scale=5, size=x.size)

y = 150 * np.sin(x) + n

sns.scatterplot(x=x, y=y);
```

Now, lets use a linear regression to predict this 

```python
import numpy as np
from sklearn.linear_model import LinearRegression

```

```python
model = LinearRegression().fit(np.array(x).reshape(-1, 1), y)
```

```python
y_pred2 = model.predict(np.array(x).reshape(-1, 1))
sns.scatterplot(x=x, y=y);
sns.regplot(x=x, y=y_pred2,scatter_kws={"color": "green"}, line_kws={"color": "red"})
```

as we can see, the model isn't good fitting this data. A straight line can't be curved to capture the true relationship. That is, it has a high bias.
To reduce bias, lets try a random forest regressor.

```python
from sklearn.ensemble import RandomForestRegressor
regr = RandomForestRegressor(bootstrap=False, random_state=0)
```

```python
model = regr.fit(np.array(x).reshape(-1, 1), y)

y_pred2 = model.predict(np.array(x).reshape(-1, 1))
sns.scatterplot(x=x, y=y_pred2);
sns.scatterplot(x=x, y=y);
```

Wow! It is a PERFECT FIT! We can't see the blue dots, but they are there. We have no bias!!! **It smells bad, right?** 

We have some problems here:
1. We are training and testing our model with the same dataset
2. We aren't able to say how the model will perform in a real dataset.
3. We have the famous Overffiting problem
